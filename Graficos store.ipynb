```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df_stores = pd.read_csv('stores.csv')
df_stores.head()
```

```python
store_table = df_stores['store_segment'].value_counts(normalize=True)
store_table
```

```python
sns.barplot(x=store_table.index, y=store_table.values, palette="Set3")
plt.title('Distribuci√≥n por store', fontsize=16)
plt.xlabel('Store', fontsize=10)
plt.xticks(rotation=90, fontsize=8)
plt.yticks(fontsize=8)
plt.show()
```

```python
store_table.plot.pie(autopct='%1.1f%%', startangle=160, colors=sns.color_palette("Set3"))
plt.title('Proporciones Stores', fontsize=16)
```
