```python
# Limpiamos el dataset de metrics
# Eliminamos duplicados
df_metrics.drop_duplicates(inplace=True)

# Comprobamos que se hayan eliminado los duplicados
df_metrics.duplicated().sum()
```

```python
# Crear máscaras para NaN y 0
mascara_nan = df_metrics.isna()  # Máscara para NaN
mascara_cero = df_metrics == 0  # Máscara para valores 0

# Combinamos ambas máscaras (NaN o 0) para cada celda
mascara_combinada = mascara_nan | mascara_cero

# Contamos el número de NaN o 0 en cada fila
conteo_nan_o_cero_por_fila = mascara_combinada.sum(axis=1)

# El umbral es el 50% de las columnas
umbral = 0.5 * df_metrics.shape[1]

# Filtrar filas con más del 50% de NaN o 0
filas_a_conservar = conteo_nan_o_cero_por_fila <= umbral

# Crear el DataFrame limpio
metrics_df_limpio = df_metrics[filas_a_conservar]

# Verificación: mostrar cuántas filas y columnas hay antes y después de la limpieza
print(f"Antes de la limpieza: {df_metrics.shape[0]} filas y {df_metrics.shape[1]} columnas.")
print(f"Después de la limpieza: {metrics_df_limpio.shape[0]} filas y {metrics_df_limpio.shape[1]} columnas.")

# Ver las primeras filas del DataFrame limpio
print("\nPrimeras filas del DataFrame limpio:")
metrics_df_limpio.head()
```

```python
# Sustituimos valores nulos por la moda o la media según corresponda

# Para columnas numéricas, sustituimos por la media
numeric_columns = metrics_df_limpio.select_dtypes(include=['float64', 'int64']).columns
metrics_df_limpio[numeric_columns] = metrics_df_limpio[numeric_columns].apply(lambda x: x.fillna(x.mean()))

# Para columnas categóricas, sustituimos por la moda
categorical_columns = metrics_df_limpio.select_dtypes(include=['object']).columns
metrics_df_limpio[categorical_columns] = metrics_df_limpio[categorical_columns].apply(lambda x: x.fillna(x.mode()[0]))

# Verificamos que no queden valores nulos
metrics_df_limpio.isnull().sum()
```

```python
# Guardamos el datest limpio
metrics_df_limpio.to_csv('metrics_clean.csv', index=False)
```
